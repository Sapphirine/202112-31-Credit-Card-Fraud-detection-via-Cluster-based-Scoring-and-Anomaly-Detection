```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# Data Loading

```python
#Reading .csv
df = pd.read_csv('creditcard.csv')

#Separating feature dimensions from class labels
X = df.iloc[:, :30]
y = df.iloc[:, 30].values

X = X.astype('float')
y = y.astype('float')
```

# Data Description

```python
#Generating a summary of statistical measures for each feature dimension, e.g., mean, count, std, min-max, etc.
X.describe()
```

# Variance & Standard-deviation captured by different principal-components of the dataset

```python
#Generating Variance of each PCA component
print("Variance:")
X.iloc[:, 1:29].var(axis = 0)
```

```python
#Generating Standard-deviation of each PCA Component
print("Standard-deviation:")
X.iloc[:, 1:29].std(axis = 0)
```

# Feature-wise histograms

```python
plt.figure()
X.hist(figsize=(30,30))
plt.show()
```

# Class-wise frequency & Outlier Fraction

```python
#Computing Outlier Fraction
print("Total fraudulent transactions (True Positives): " + str(len(y[y == 1])))
print("Total non-fraudulent transactions (True Positives): " + str(len(y[y == 0])))
print("")
print("Outlier Fraction (n(fraudulent)/n(non-fraudulent)): " + str(len(y[y == 1])/len(y[y == 0])))
```

# Correlation heatmap

```python
#Generating correlation among feature dimenions and plotting a heatmap
plt.figure(figsize=(10, 10))
sns.heatmap(X.corr(), square = True)
plt.show()
```
