```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import collections

from sklearn.metrics import confusion_matrix, auc
```

# Data Loading

```python
#Reading .csv
df = pd.read_csv('creditcard.csv', header = None)

#Separating feature dimensions from calss labels
X_initial = df.iloc[1:, 1:30].values
y = df.iloc[1:, 30].values

X_initial = X_initial.astype('float')
y = y.astype('float')
```

# Hyper Parameters

```python
#Initializing the list of K-values from the set [2 -> 20]
K = list(range(2, 21))

#Number of bootstrapping runs
runs = 10

#Number of iterations for k-means algorithm
kmeans_max_iterations = 500

#Number of times centroids are initialized per k-means training phase to find the best centroid initialization
n_init_parameter = 10
```

# Initializing Variables & Dictionaries

```python
n = len(X_initial)
N = len(K)

#Initializing dictionaries to store results for each run with the format {data[run_number]: [results]}
score = {}
y_true = {}
y_preds = {}
auprc = {}
auroc = {}
```

# Proposed Algorithm & Generating PRC, ROC Curves, Scatter Plots (Qualitative Analysis)

```python
#Starting the loop for bootstrapping runs
for r in range(runs):
    print("")
    print("Run " + str(r+1))
    
    #Generating n random indexes with replacement
    bootstrap_index = np.random.choice(n, size = n, replace = True)
    
    #Selecting bootstrapped data (features & respective labels) using generated indexes
    X = X_initial[bootstrap_index]
    
    X = (X - X.mean(axis = 0)) / (X.max(axis = 0) - X.min(axis = 0))
    
    y_true[r+1] = y[bootstrap_index]
    
    #Initializing the numerator and denomenator arrays for calculating consistency score
    score_num = np.zeros(n)
    score_denom = np.zeros(n)
    
    #Initializing dictionaries to store centroids, centroid indexes, and number of elements in each centroid
    C = {}
    C_index = {}
    C_freqs = {}
    
    #Starting for loop to perform K-means with K different values
    for k in K:
    
        print("Running k-means for k = " + str(k))
        
        #Initializing Sklearn's Kmeans Class with hyperparameters defined above
        kmeans = KMeans(n_clusters = k, init='k-means++', max_iter = kmeans_max_iterations, n_init = n_init_parameter, random_state = 0)
        
        #Begin training process
        cluster_index = kmeans.fit_predict(X)
        
        #Generating list of assigned clusters in the form of indexes
        cluster_centers = kmeans.cluster_centers_ 
        
        #Generating centroids using indexes for each data point
        clusters = [cluster_centers[i] for i in cluster_index]
        clusters = np.array(clusters)
    
        #Storing centroids and the respective indexes in the dictionaries initialized above
        C[k] = clusters
        C_index[k] = cluster_index
    
    #Starting for loop to compute number of data points assigned to each cluster
    for k in K:
        print("Calculating frequencies for k = " + str(k))
        
        #Generating a dictionary of counts for each cluster index using collections.Counter() function
        freq = dict(collections.Counter(C_index[k]))
        temp = np.zeros(n)
        for i in range(n):
            #Computing the total sum of cluster counts for each datapoint
            temp[i] = freq[C_index[k][i]]
        
        #Storing the final result in the dictionary C_freqs with key k
        C_freqs[k] = temp
    
    #Starting double for loop to calculate double summation of consistency score
    for i in range(N):
        for j in range(i+1, N):
            print("Calculating weighted score using Centroids: " + str(i+1) + " & " + str(j+1))
            
            #Computing cosine similarity between data points for k = i,j
            cos = np.sum(np.multiply(C[K[i]], C[K[j]]), axis = 1) / np.multiply(np.linalg.norm(C[K[i]], ord = 2, axis = 1), np.linalg.norm(C[K[j]], ord = 2, axis = 1))
            
            #Computing numerator of similarity score
            score_num += (C_freqs[K[i]] + C_freqs[K[j]]) * cos
            
            #Computing denomenator of similarity score
            score_denom += (C_freqs[K[i]] + C_freqs[K[j]])
    
    #Computing the similarity score
    score[r+1] = score_num / score_denom
    
    #Plotting histogram of generated scores
    plt.figure()
    plt.title('Score distribution for run ' + str(r+1))
    plt.hist(score[r+1])
    plt.grid()
    plt.xlabel('Score')
    plt.ylabel('Frequency')
    plt.show()
    
    #Initializing a set of 1000 thresholds from the set (min(score), max(score))
    print("Plotting Precision-Recall and ROC Curves for run " + str(r+1))
    thresholds = np.linspace(score[r+1].min(), score[r+1].max(), num = 1001)[1:]

    #Initializing list of results for current run
    y_preds[r+1] = np.zeros(n) #predicted labels
    recall = np.zeros(len(thresholds)) #recall values
    precision = np.zeros(len(thresholds)) #precision values
    fpr = np.zeros(len(thresholds)) #false positive rate
    
    #Looping through each treshold
    for t in range(len(thresholds)):
        #print("Run " + str(r+1) + ", threshold " + str(t+1))
        
        #Set label = 1 if predicted score is below threshold
        y_preds[r+1][np.where(score[r+1] < thresholds[t])] = 1
        
        #Using Sklearn's confusion_matrix function to generate tn, fp, fn, tp values
        tn, fp, fn, tp = confusion_matrix(y_true[r+1], y_preds[r+1]).ravel()
        
        #Computing recall, precision and false positive rate
        recall[t] = tp/(tp + fn)
        precision[t] = tp/(tp + fp)
        fpr[t] = fp/(fp + tn)
    
    #Calculating area under the precision-recall curve
    i = np.array(range(1, len(recall)))
    re = recall[i] - recall[i-1]
    pre = precision[i] + precision[i-1]
    auprc[r+1] = np.sum(re * pre)/2
    
    #Plotting Precision-Recall curve
    plt.figure()
    plt.title('Precision-Recall Curve for run ' + str(r+1))
    plt.plot(recall, precision, 'b', label = 'AUPRC = %0.6f' % auprc[r+1])
    plt.grid()
    plt.legend(loc = 'upper right')
    plt.show()
    
    #Calculating area under the recall curve using Sklearn's auc function
    auroc[r+1] = auc(fpr, recall) 
    
    #Plotting ROC Curve
    plt.figure()
    plt.title('Receiver Operating Characteristic for run ' + str(r+1))
    plt.plot(fpr, recall, 'b', label = 'AUROC = %0.6f' % auroc[r+1])
    plt.grid()
    plt.legend(loc = 'lower right')
    plt.show()
```

```python
#Initializing a list of thresholds to generate scatter plots between pairs of PCA components
scatter_plot_thresholds = [0.1, 0.2, 0.3]

#Starting for loop to generate plots for 6 different pairs
for i in range(6):
    
    #Randomly choosing 2 PCA components from a set of 27, p1 and p2
    p1, p2 = np.random.randint(0, 27, 2)
    print("Generating scatter plots for V" + str(p1+1) + " vs. V" + str(p2+1))
    
    #Assigning vectors with indexes p1 and p2 as t1 and t2 respectively
    t1 = X[:, p1]
    t2 = X[:, p2]
    
    #Starting for loop across each threshold value
    for t in scatter_plot_thresholds:
        
        #Generating predicted labels based on score values
        yo = np.zeros(n)
        s = score[1]
        yo[np.where(s < t)] = 1
        y = y_true[1]
    
        o = np.empty(n, dtype = object)
        
        #Assigning output as tp, fp, tn, fn for each data point
        for i in range(n):
            if yo[i] == y[i] == 1:
                o[i] = 'tp'

            if yo[i] == 1 and yo[i] != y[i]:
                o[i] = 'fp'

            if yo[i] == y[i] == 0:
                o[i] = 'tn'

            if yo[i] == 0 and yo[i] != y[i]:
                o[i] = 'fn'
        
        #Plotting scatter plot between t1 and t2 with tp, fp, tn, fn as labels
        plt.figure()
        plt.title("Scatter plot of V" + str(p1+1) + " vs. V" + str(p2+1) + " for threshold = " + str(t))
        plt.scatter(t1[o == 'tn'], t2[o == 'tn'], label = 'tn', s = 5, c = "g")
        plt.scatter(t1[o == 'fp'], t2[o == 'fp'], label = 'fp', s = 5, c = "b")
        plt.scatter(t1[o == 'fn'], t2[o == 'fn'], label = 'fn', s = 5, c = "c")
        plt.scatter(t1[o == 'tp'], t2[o == 'tp'], label = 'tp', s = 5, c = "r")
        plt.legend()
        plt.xlabel("V" + str(p1+1))
        plt.ylabel("V" + str(p2+1))
        plt.show()
```

```python
#Initializing list of thresholds to display tp and tn
table_thresholds = [0.5, 0.6, 0.7, 0.8, 0.9]

#Initializing dictionary to store results
table_results = {}

#Starting for loop across each treshold value
for t in table_thresholds:
    
    #Generating predicted labels based on scores
    yo = np.zeros(n)
    s = score[1]
    yo[np.where(s < t)] = 1
    y = y_true[1]

    tp = 0
    tn = 0
    
    #For each data point, computing the total number of tp and tn
    for i in range(n):
        if yo[i] == y[i] == 1:
            tp += 1
            
        if yo[i] == y[i] == 0:
            tn += 1
    
    #Storing the results in total_results dictionary
    table_results[">" + str(t)] = {'True Positives': tp, 'True Negatives': tn}

#Displaying the result for each threshold
print("Number of TP's and TN's for higher consistency scores thresholds:\n")
for k, v in table_results.items():
    print(k + " :" + str(v))
```
